# Bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x = ['Java','C++','Python','C#']
y = [80,70,90,60]
z = [90,80,80,80]
plt.xlabel("Langurages")
plt.ylabel("Rating")
plt.title("languages with rating")
col = ['y','b','m','g']
#plt.bar(x,y,width=0.4,color = col,align='edge',edgecolor='black',linewidth=5,linestyle=':',alpha=0.5)
plt.bar(x,y,width=0.4,align='edge',edgecolor='black',label='popularity in india')
plt.bar(x,z,width=0.4,label="popularity in pakistan")
plt.legend()
plt.show()
```

# scatter plot


```python
x = [1,2,3,4,5,6]
y = [30,40,35,23,26,10]


plt.title("Scatter plot")
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.scatter(x,y,color='r',s=100,marker="*")
#plt.colorbar()
plt.show()
```

# Histogram 

```python
x = np.random.randint(10,100,50)
print(x)
```

```python
x = """17 58 25 17 77 96 72 31 74 58 64 57 95 29 83 39 70 12 17 35 31 33 77 28
 73 90 13 28 58 51 36 43 82 19 71 55 22 62 62 27 38 79 14 16 22 23 59 28
 35 59"""
x = x.split()
y = []
for i in x:
    y.append(int(i))
x = y.copy()
```

```python
plt.hist(x)
plt.title("histogram")
plt.xlabel("Python")
plt.ylabel("Numbers")
plt.hist(x,edgecolor='r',color='g',cumulative=-1,rwidth=0.9)
#plt.xlim(10,100)
plt.axvline(45,color='b',label="current target")
plt.legend()
plt.show()
```

# pie chart

```python
x=[10,20,30,40]

y = ['c++','python','c','c#']
ex = [0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.0f%%",shadow=True,radius=1.3,rotatelabels=True)
plt.legend(loc=2)
plt.show()
```

```python
x=[10,20,30,40]

y = ['c++','python','c','c#']
ex = [0.4,0.0,0.0,0.0]
c = ['r','g','b','y']
plt.pie(x,labels=y,colors=c,radius=1.5)
plt.pie([1],colors='w')
plt.show()
```

# stem plot

```python
x = [2,4,5,1,8,6,5,6]
y = [9,6,5,4,6,4,2,4]

plt.stem(x,y,linefmt=":",markerfmt="y*",bottom=1,basefmt="g",label="python ratings",orientation="horizontal")
plt.xlabel("Data")
plt.ylabel("Ratings")
plt.legend()
plt.show()
```

# Box plot 

```python
x = [10,20,30,40,50,60,70,150]
x1 = [10,30,60,40,120,80,120]
#plt.boxplot(x,notch=False)
#plt.boxplot(x,vert=False)
y = [x,x1]
plt.boxplot(y,widths=0.5,labels=['python','c++'],patch_artist=True,showmeans=True,whis=2,sym='g+')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,7,3,8,4]
y1 = [12,23,12,14,15]
y2 = [20,23,30,40,50]

label = ["area1","area2","area3"]
plt.stackplot(x,y,y1,y2,labels=label)
plt.legend(loc=2)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.title("matplotlib")
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.step(x,y,color="r",marker="o",ms=10,mfc="g")
plt.grid()
plt.show()
```

# filling

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.plot(x,y,color="r")
plt.show()
```

# sub-plot
