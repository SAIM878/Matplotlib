# Bar plot

```python
import matplotlib.pyplot as plt
```

```python
x = ['Java','C++','Python','C#']
y = [80,70,90,60]
z = [90,80,80,80]
plt.xlabel("Langurages")
plt.ylabel("Rating")
plt.title("languages with rating")
col = ['y','b','m','g']
#plt.bar(x,y,width=0.4,color = col,align='edge',edgecolor='black',linewidth=5,linestyle=':',alpha=0.5)
plt.bar(x,y,width=0.4,color = col,align='edge',edgecolor='black',label='popularity in india')
plt.bar(x,z,width=0.4,label="popularity in pakistan")
plt.legend()
plt.show()
```
